# Upstox Portfolio Analysis
This Jupyter Notebook connects to your Upstox account to fetch your live holdings and positions. You can use it to read your data into Pandas and prompt your AI assistant to analyze it.

```python
import json
import pandas as pd
from broker.upstox_api import UpstoxAPI

# Load Upstox credentials
with open('config.json', 'r') as f:
    token = json.load(f).get('access_token')

client = UpstoxAPI(access_token=token)
print("Upstox Client Initialized!")
```

```python
# Fetch Holdings (Long-term)
try:
    holdings_response = client._request('GET', '/portfolio/long-term-holdings')
    holdings = holdings_response.get('data', [])
    if holdings:
        df_holdings = pd.DataFrame(holdings)
        print(f"Found {len(df_holdings)} holdings.")
        display(df_holdings)
    else:
        print("No long-term holdings found.")
except Exception as e:
    print(f"Error fetching holdings: {e}")
```

```python
# Fetch Positions (Short-term)
try:
    positions_response = client._request('GET', '/portfolio/short-term-positions')
    positions = positions_response.get('data', [])
    if positions:
        df_positions = pd.DataFrame(positions)
        print(f"Found {len(df_positions)} positions.")
        display(df_positions)
    else:
        print("No short-term positions found.")
except Exception as e:
    print(f"Error fetching positions: {e}")
```
